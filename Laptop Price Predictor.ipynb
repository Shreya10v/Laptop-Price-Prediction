### Model Building

```python
# 1. LabelEncoder - 
# 2. train test split
# 3. building,fit,predict with models
# 4. for each model , R2 score and MAE
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler,StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import LabelEncoder,OneHotEncoder
from sklearn import metrics
from sklearn.model_selection import RandomizedSearchCV
from sklearn.linear_model import LinearRegression,Lasso,Ridge
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor,GradientBoostingRegressor
from sklearn.svm import SVR
from sklearn.neighbors import KNeighborsRegressor
from sklearn import tree
from xgboost import XGBRegressor
import sys
print(sys.executable)
```

```python
test = np.log(df['Price'])
train = df.drop(['Price'],axis = 1)
```

```python
X_train, X_test, y_train, y_test = train_test_split(train,test,
                                                   test_size=0.15,random_state=2)

X_train.shape,X_test.shape
```

```python
#mapping column names to their index no. to be used in Pipelines in next steps
mapper = {i:value for i,value in enumerate(X_train.columns)}
mapper
```

```python
models_to_be_tested = [LinearRegression(), Ridge(alpha=10), Lasso(alpha=0.001), DecisionTreeRegressor(max_depth=8), RandomForestRegressor(n_estimators=100,
                              random_state=3,
                              max_samples=0.5,
                              max_features=0.75,
                              max_depth=15)]
len(models_to_be_tested)
```

```python
for i in models_to_be_tested:
    step1 = ColumnTransformer(transformers=[('col_tnf',OneHotEncoder(drop='first'),[0,1,3,8,11])],remainder='passthrough')
    
    step2 = i
    
    pipe = Pipeline([('step1',step1), ('step2',step2)])
    
    pipe.fit(X_train,y_train)
    
    y_pred = pipe.predict(X_test)

    print(f"Evaluation metrics for model - {i}")
    print('R2 score',metrics.r2_score(y_test,y_pred))
    print('MAE',metrics.mean_absolute_error(y_test,y_pred))
```

## Since Random Forest has highest R2 score, selecting it as our prediction model

```python
step1 = ColumnTransformer(transformers=[
    ('col_tnf',OneHotEncoder(drop='first'),[0,1,3,8,11])
],remainder='passthrough')

step2 = RandomForestRegressor(n_estimators=100,
                              random_state=3,
                              max_samples=0.5,
                              max_features=0.75,
                              max_depth=15)

pipe = Pipeline([
    ('step1',step1),
    ('step2',step2)
])

pipe.fit(X_train,y_train)

y_pred = pipe.predict(X_test)

print('R2 score',metrics.r2_score(y_test,y_pred))
print('MAE',metrics.mean_absolute_error(y_test,y_pred))
```

## label encoding our categorical columns to save  the selected model

```python
indexlist = [0,1,3,8,11]
transformlist = []
for key,value in mapper.items():
    if key in indexlist:
        transformlist.append(value)
        
transformlist
```

```python
train = pd.get_dummies(train,columns=transformlist,drop_first=True)
train.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(train,test,
                                                   test_size=0.15,random_state=2)

X_train.shape,X_test.shape
```

```python
# saving the model
rf1 = RandomForestRegressor(n_estimators=100,
                              random_state=3,
                              max_samples=0.5,
                              max_features=0.75,
                              max_depth=15)

rf1.fit(X_train,y_train)
print(f'R2 score : {metrics.r2_score(y_test,rf1.predict(X_test))}')
```
